```python
df = merged
# Load your DataFrame (replace with your data loading method)
# df = pd.read_csv('your_file.csv')

# List of prediction columns including 'compas_guess'
h_columns = [f'h{i}' for i in range(1, 101)] + ['compas_guess']

# Initialize dictionaries to hold metrics for each sex group
metrics_sex_0 = {h: {'TP': 0, 'FP': 0, 'FN': 0, 'TN': 0} for h in h_columns}
metrics_sex_1 = {h: {'TP': 0, 'FP': 0, 'FN': 0, 'TN': 0} for h in h_columns}

# Calculate metrics for sex = 0
df_sex_0 = df[df['sex'] == 0]
for h in h_columns:
    tp = ((df_sex_0[h] == 1) & (df_sex_0['two_year_recid'] == 1)).sum()
    fp = ((df_sex_0[h] == 1) & (df_sex_0['two_year_recid'] == 0)).sum()
    fn = ((df_sex_0[h] == 0) & (df_sex_0['two_year_recid'] == 1)).sum()
    tn = ((df_sex_0[h] == 0) & (df_sex_0['two_year_recid'] == 0)).sum()
    
    metrics_sex_0[h]['TP'] = tp
    metrics_sex_0[h]['FP'] = fp
    metrics_sex_0[h]['FN'] = fn
    metrics_sex_0[h]['TN'] = tn

# Calculate metrics for sex = 1
df_sex_1 = df[df['sex'] == 1]
for h in h_columns:
    tp = ((df_sex_1[h] == 1) & (df_sex_1['two_year_recid'] == 1)).sum()
    fp = ((df_sex_1[h] == 1) & (df_sex_1['two_year_recid'] == 0)).sum()
    fn = ((df_sex_1[h] == 0) & (df_sex_1['two_year_recid'] == 1)).sum()
    tn = ((df_sex_1[h] == 0) & (df_sex_1['two_year_recid'] == 0)).sum()
    
    metrics_sex_1[h]['TP'] = tp
    metrics_sex_1[h]['FP'] = fp
    metrics_sex_1[h]['FN'] = fn
    metrics_sex_1[h]['TN'] = tn

# Convert the metrics dictionaries to DataFrames
metrics_df_sex_0 = pd.DataFrame(metrics_sex_0).T
metrics_df_sex_1 = pd.DataFrame(metrics_sex_1).T

# Show the metrics DataFrames
print("Metrics for sex = 0:")
print(metrics_df_sex_0)

print("\nMetrics for sex = 1:")
print(metrics_df_sex_1)

# Optionally save the metrics DataFrames to CSV
# metrics_df_sex_0.to_csv('metrics_sex_0.csv')
# metrics_df_sex_1.to_csv('metrics_sex_1.csv')
```

```python
df = merged

h_columns = [f'h{i}' for i in range(1, 21)] + ['compas_guess']

# Initialize a dictionary to hold metrics for each prediction column
metrics = {h: {'TP': 0, 'FP': 0, 'FN': 0, 'TN': 0} for h in h_columns}

# Calculate TP, FP, FN, TN
for h in h_columns:
    # True Positives (TP): Predicting 1 when true label is 1
    tp = ((df[h] == 1) & (df['two_year_recid'] == 1)).sum()
    # False Positives (FP): Predicting 1 when true label is 0
    fp = ((df[h] == 1) & (df['two_year_recid'] == 0)).sum()
    # False Negatives (FN): Predicting 0 when true label is 1
    fn = ((df[h] == 0) & (df['two_year_recid'] == 1)).sum()
    # True Negatives (TN): Predicting 0 when true label is 0
    tn = ((df[h] == 0) & (df['two_year_recid'] == 0)).sum()

    # Store the calculated values in the metrics dictionary
    metrics[h]['TP'] = tp
    metrics[h]['FP'] = fp
    metrics[h]['FN'] = fn
    metrics[h]['TN'] = tn

# Convert the metrics dictionary to a DataFrame for better readability
metrics_df = pd.DataFrame(metrics).T

# Show the metrics DataFrame
metrics_df

# Optionally save the metrics DataFrame to a CSV
# metrics_df.to_csv('metrics.csv')
```

```python
def independence(tp, fp, fn, tn):
    return (tp+fp)/(tp+fp+fn+tn)
def specificity(tp, fp, fn, tn):
    return tn/(tn+fp)
def recall(tp, fp, fn, tn):
    return tp/(tp+fn)
def precision(tp, fp, fn, tn):
    return tp/(tp+fp)
def sufficiency(tp, fp, fn, tn):
    return tn/(tn+fn)
```

```python
metrics_df_sex_0['independence'] = metrics_df_sex_0.apply(lambda row: independence(row['TP'], row['FP'], row['FN'], row['TN']), axis=1)
metrics_df_sex_0['specificity'] = metrics_df_sex_0.apply(lambda row: specificity(row['TP'], row['FP'], row['FN'], row['TN']), axis=1)
metrics_df_sex_0['recall'] = metrics_df_sex_0.apply(lambda row: recall(row['TP'], row['FP'], row['FN'], row['TN']), axis=1)
metrics_df_sex_0['precision'] = metrics_df_sex_0.apply(lambda row: precision(row['TP'], row['FP'], row['FN'], row['TN']), axis=1)
metrics_df_sex_0['sufficiency'] = metrics_df_sex_0.apply(lambda row: sufficiency(row['TP'], row['FP'], row['FN'], row['TN']), axis=1)
metrics_df_sex_0
```
