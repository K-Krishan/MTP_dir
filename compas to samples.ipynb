```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
votes = pd.read_csv('MTURK_RACE.csv')
df = pd.read_csv('BROWARD_CLEAN.csv')
```

```python
votes.head()
```

```python
df.head()
```

```python
counts = pd.DataFrame()
counts['mTurk_code'] = votes['mTurk_code']
counts['yes'] = votes.apply(lambda row : sum(row[1:] == 1), axis=1)
counts['no'] = votes.apply(lambda row : sum(row[1:] == -1), axis=1)
counts['total'] = (-votes.isnull()).sum(axis=1)
```

```python
counts.head()
```

```python
distribution = pd.DataFrame()
distribution['mTurk_code'] = counts['mTurk_code']
distribution['yes'] = counts['yes'] / counts['total']
distribution['no'] = counts['no'] / counts['total']
```
